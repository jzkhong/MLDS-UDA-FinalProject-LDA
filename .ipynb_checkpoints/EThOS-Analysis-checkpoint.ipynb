```python
def generate_word_cloud(text, max_words=100, random_state=32, background_color="white", width=1600, height=800):
    """
    Generates a word cloud from a given text string.

    This function creates a word cloud visualization for the provided text.

    Parameters:
    text (str): The text string to generate the word cloud from.
    max_words (int): The maximum number of words to display in the word cloud. Default is 100.
    random_state (int): Seed for random number generator for reproducibility. Default is 32.
    background_color (str): Background color for the word cloud image. Default is "white".
    width (int): Width of the word cloud image. Default is 1200.
    height (int): Height of the word cloud image. Default is 600.

    Returns:
    WordCloud: The generated word cloud object.
    """
    # Generate a word cloud
    word_cloud = WordCloud(max_words=max_words,
                           random_state=random_state, 
                           stopwords=ENGLISH_STOP_WORDS, # exclude stop words
                           normalize_plurals=False, # disable automatic plural handling
                           background_color=background_color,
                           # adjust width and height for better resolution
                           width=width, height=height)
    word_cloud.generate(text)

    return word_cloud
```

## Check for non-English title

```python
# Define a corpus of English words
english_corpus = set()
english_corpus.update(words.words())
english_corpus.update(wordnet.words())

# Add additional english terms to the corpus
additional_terms = {'containing','crystallising','comprehending','lloyds','multichannel','nanostructures',
                    'nanoelectronics','photoelectrochemistry','signalling','superconducting','targeting',
                    'translating','william'}
english_corpus.update(additional_terms)

# Define the lemmatizer
lemmatizer = WordNetLemmatizer()

def is_english(text, lemmatizer, english_corpus):
    """
    Checks if the majority of words in a given text are English.

    Processes the input text by lowercasing, removing punctuation, tokenizing, and lemmatizing. 
    It then compares the lemmatized words against a predefined set of English words ('english_corpus'). 
    The function returns True if at least 30% of the words are found in the English corpus.

    Parameters:
    text (str): The text to be analyzed.
    lemmatizer (nltk.stem.WordNetLemmatizer): The lemmatizer to use for processing the text.
    english_corpus (set): A set of English words to compare against.

    Returns:
    bool: True if >=30% of words are English, False otherwise.
    """
    # lowercase text, remove punctuations/symbols, and split text into words
    words_in_text = set(nltk.word_tokenize(re.sub(r"[^\w\s]", " ", text.lower())))
    # perform lemmatization
    lemma_in_text = [lemmatizer.lemmatize(token) for token in words_in_text]
    # count the number of tokens exists in the English corpus
    english_word_count = sum(word in english_corpus for word in lemma_in_text)

    # return True if more than 30% of words in text are in the english corpus defined
    return english_word_count / len(words_in_text) >= 0.30

# Identify non-english title
data['is_english'] = data['Title'].apply(lambda x: is_english(x, lemmatizer, english_corpus))
print(data['is_english'].value_counts())
```

# Pre-processing Thesis Title

## Stopwords Removal & Lemmatization

```python
# Filter out non-english title
eng_data = data[(data['is_english']==True)].copy()
print("There are {:,} English title.".format(len(eng_data)))
```

```python
# Define stop words set
stop_words = ENGLISH_STOP_WORDS

# Add corpus related stop words
stop_words = stop_words.union({'study', 'development', 'analysis', 'investigation', 'effect', 'role'})

def preprocessing(text, lemmatizer, stop_words):
    """
    Processes and cleans the input text for natural language processing tasks.

    The function converts text to lowercase, removes apostrophe-s, single-letter abbreviations, 
    and punctuations. It tokenizes the text, lemmatizes tokens, and excludes stop words 
    and non-alphabetical tokens.

    Parameters:
    text (str): The text to preprocess.
    lemmatizer (nltk.stem.WordNetLemmatizer): The lemmatizer to use for processing the text.
    stop_words (set): Set of words to be excluded.

    Returns:
    list: List of lemmatized, non-stop word tokens from the text.
    """
    # lowercase and remove apostrophe-s at end of words
    text = re.sub("'s\\b", " ", text.lower())
    # replace single letters followed by a period with a space
    text = re.sub(r"\b[a-zA-Z]\.", " ", text)
    # remove punctuation (any character that is not a word character)
    text = re.sub(r"[^\w\s]"," ", text)
    # split text into words
    tokens = nltk.word_tokenize(text)
    # perform lemmatization on words with only alphabets and remove stop words
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha() 
              and lemmatizer.lemmatize(token) not in stop_words]
    return tokens

# Preprocess title using the WordNet Lemmatizer defined during checks for non-English title
data_words = eng_data['Title'].apply(lambda x: preprocessing(x, lemmatizer, stop_words))

# Add processed title in data dataframe
eng_data['processed_title'] = [' '.join(doc) for doc in data_words]

# Review a few examples after cleaning thesis title 
print('Before: {}\nAfter: {}\n'.format(eng_data['Title'][1],eng_data['processed_title'][1]))
print('Before: {}\nAfter: {}\n'.format(eng_data['Title'][37],eng_data['processed_title'][37]))
print('Before: {}\nAfter: {}\n'.format(eng_data['Title'][87],eng_data['processed_title'][87]))
```

# TF-IDF Analysis

```python
def get_top_tfidf_keywords(data, top_n=10):
    """
    Calculate and return the top N keywords by TF-IDF score from a collection of documents.
    
    Parameters:
    - data: Iterable of text documents.
    - top_n (int): Number of top keywords to return. Defaults to 10.
    
    Returns:
    - DataFrame containing the 'term' and its 'score', sorted by score in descending order.
    """
    # Define the TfidfVectorizer
    tfidf_vectorizer = TfidfVectorizer(sublinear_tf=True, # Apply sublinear scaling: 1+log(tf) 
                                       max_df=0.5 # Exclude terms with a document frequency more than 50%
                                      )
    # Apply vectorizer to transform data
    tfidf_matrix = tfidf_vectorizer.fit_transform(data)
    # Extract results into a pandas dataframe
    feature_names = tfidf_vectorizer.get_feature_names_out()
    tfidf_scores = tfidf_matrix.sum(axis=0).A1
    scores_df = pd.DataFrame({'term': feature_names, 'score': tfidf_scores})
    return scores_df.sort_values('score', ascending=False).head(top_n)
```

```python
# Get the top 10 terms based on TF-IDF for the 20th and 21st centuries
twentieth_century_tfidf = get_top_tfidf_keywords(eng_data[eng_data['Date']<=2000]['processed_title'], 10)
twenty_first_century_tfidf = get_top_tfidf_keywords(eng_data[eng_data['Date']>2000]['processed_title'], 10)

# Extract the titles from the top 2 disciplines for the 20th and 21st centuries
twentieth_century_top_discipline_titles = eng_data[(eng_data['Date']<=2000) 
& (eng_data['Subject Discipline'].isin(['Medicine & Health','Physical Sciences']))]['processed_title']
twenty_first_century_top_discipline_titles = eng_data[(eng_data['Date']>2000) 
& (eng_data['Subject Discipline'].isin(['Medicine & Health','Physical Sciences']))]['processed_title']

# Get the top 10 terms from the top 2 disciplines based on TF-IDF for the 20th and 21st centuries
twentieth_century_top_discipline_tfidf = get_top_tfidf_keywords(twentieth_century_top_discipline_titles, 10)
twenty_first_century_top_discipline_tfidf = get_top_tfidf_keywords(twenty_first_century_top_discipline_titles, 10)

# Assign a category to each DataFrame
twentieth_century_tfidf['Category'] = '20th Century Overall'
twenty_first_century_tfidf['Category'] = '21st Century Overall'
twentieth_century_top_discipline_tfidf['Category'] = '20th Century Top Two Disciplines'
twenty_first_century_top_discipline_tfidf['Category'] = '21st Century Top Two Disciplines'

# Combine the TF-IDF into a single data frame
tfidf_results = pd.concat([
    twentieth_century_tfidf, 
    twenty_first_century_tfidf, 
    twentieth_century_top_discipline_tfidf, 
    twenty_first_century_top_discipline_tfidf
], axis=0)
```

# Bigram Generation

```python
# Build the bigram models
bigram = gensim.models.Phrases(data_words, min_count=10)
bigram_mod = gensim.models.phrases.Phraser(bigram)

# Generate bigrams
def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

data_words_bigrams = make_bigrams(data_words)

# Examine some of the bigrams
for idx, doc in enumerate(data_words_bigrams[:5]):  # Look at the first 5 documents
    print(f"Thesis {idx+1} bigrams:")
    print([word for word in doc if '_' in word])  # Print only the bigrams
```

** **
# Latent Dirichlet Allocation (LDA)

```python
# Create a Dictionary
common_dictionary = Dictionary(data_words_bigrams)

# Filter out extremes terms
# no_below: minimum number of documents in which a word must appear
# no_above: maximum proportion of documents in which a word can appear (e.g., 0.5 means a word cannot be in more than 50% of the documents)
# keep_n: keep only the first `keep_n` most frequent tokens
common_dictionary.filter_extremes(no_below=10, no_above=0.3, keep_n=100000)

# Convert data to vectors
common_corpus = [common_dictionary.doc2bow(doc) for doc in data_words_bigrams]
```

```python
print('Unique tokens in dictionary:', len(common_dictionary))
```

## Hyperparameters Tuning

```python
# Iterate through different configuration of number of topics
coherence_score1 = []
topics = []
for i in range(5, 16):

    # Set seed for reproducibility
    seed = 32
    random.seed(seed)
    numpy.random.seed(seed)
    k = i 

    # Train LDA model
    # Note: Gensim LdaModel package is used instead of LdaMultiCore to ensure reproducibility
    lda_model = LdaModel(corpus=common_corpus, id2word=common_dictionary,
                         num_topics=k, passes=3, random_state=seed)
    
    # Compute 'c_v' coherence score
    model_coherence = CoherenceModel(model=lda_model, texts=data_words_bigrams, 
                                     dictionary=common_dictionary, coherence='c_v')
    coherence = model_coherence.get_coherence()
    
    # Store and print the number of topics and its corresponding coherence score
    topics.append(k)
    coherence_score1.append(coherence)
    print('Topics {}: Coherence {:.4}'.format(k, coherence))
```

##### Figure 6 of PDF report: Coherence Score for Different Number of Topics (LDA)

```python
# Visualise the coherence score for different number of topics
plt.plot(topics, coherence_score1, marker='.')
plt.title('Coherence Score for Different Number of Topics (LDA)')
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Score')
plt.grid(True)
# Export visual to be used as Figure 6 of PDF report
plt.savefig('results/Fig6_LDA_No_of_Topics.pdf')
plt.show()
```

##### Table 2 of PDF report: Coherence Scores for different Alpha and Eta configurations

```python
# Iterate through different configuration of alpha and eta
coherence_score2 = []
alpha = [0.01, 0.3, 0.6, 0.9, 'symmetric', 'asymmetric', 'auto']
eta = [0.01, 0.3, 0.6, 0.9, 'symmetric', 'auto']
for a in alpha:
    for e in eta:
        
        # Set seed for reproducibility
        seed = 32
        random.seed(seed)
        numpy.random.seed(seed)
    
        # Build LDA model
        lda_model = LdaModel(corpus=common_corpus, id2word=common_dictionary,
                             num_topics=8, passes=3, random_state=seed, alpha=a, eta=e)
        # Compute coherence score
        model_coherence = CoherenceModel(model=lda_model, 
                                         texts=data_words_bigrams, 
                                         dictionary=common_dictionary, 
                                         coherence='c_v')
        coherence = model_coherence.get_coherence()
        
        # Store and print the alpha and eta configurations and its corresponding coherence score
        coherence_score2.append(coherence)
        print('Alpha {} Eta {}: Coherence {:.4}'.format(a, e, coherence))
```

## Final LDA model

```python
# Set seed for reproducibility
seed = 32
random.seed(seed)
numpy.random.seed(seed)
num_topics = 8

# Train final LDA model
lda_model = LdaModel(corpus=common_corpus, id2word=common_dictionary,
                     num_topics=num_topics, passes=20, random_state=seed, alpha=0.01, eta=0.6)
# Compute and print coherence score
model_coherence = CoherenceModel(model=lda_model, 
                                 texts=data_words_bigrams, 
                                 dictionary=common_dictionary, 
                                 coherence='c_v')
coherence = model_coherence.get_coherence()
print('Topics {}: Coherence {:.3}'.format(num_topics, coherence))
```

```python
# Group by year and calculate the distribution of topics for each year
topic_over_time = eng_data[['Date','dominant_topic']].groupby('Date')['dominant_topic'].value_counts(normalize=True).unstack()

# Plotting the heatmap topic distribution over time
plt.figure(figsize=(20, 10))
sns.heatmap(topic_over_time.T, cmap='YlOrBr', annot=False) # Transpose for topics as rows
plt.title('Topic Prevalence Over Time')
plt.xlabel('Year')
plt.ylabel('Topics')
plt.xticks(rotation=45) # Rotate the x-axis labels for better readability
plt.tight_layout()
# Export visual to be used as Figure 9 of PDF report
plt.savefig('results/Fig9_topic_evolution.pdf')
plt.show()
```

## Topic Intensity

```python
# Extract topic probabilities for each document from LDA model
topics_probs = pd.DataFrame([dict(lda_model.get_document_topics(bow, minimum_probability=0)) 
                                for bow in common_corpus])

# Add the 'Date' column (publication year) from eng_data data frame
topics_probs['Date'] = eng_data['Date']

# Compute the sum of probabilities for each topic by year
topics_by_year = topics_probs.groupby('Date').sum()

# Compute the topic intensity for each topic for each year
topics_intensity_by_year = topics_by_year.div(topics_by_year.sum(axis=1), axis=0)
```

```python
# Visualise the trend of topic intensity over time
plt.figure(figsize=(15, 7))
for topic_num in topics_intensity_by_year.columns:
    plt.plot(topics_intensity_by_year.index, topics_intensity_by_year[topic_num], 
             label=f'Topic {topic_num}')
plt.xlabel('Year')
plt.ylabel('Topic Probability')
plt.title('Trend of Topic Intensity Over Time')
# Set legend
plt.legend(loc='upper right', ncol=2, title='Topics')
plt.grid(True)
plt.show()
```

##### Figure 10 of PDF report: Topic Intensity

```python
# Apply a moving average to smooth out the data (10 years moving average)
window_size = 10
topics_intensity_smoothed = topics_intensity_by_year.rolling(window=window_size).mean()

# Visualise the trend of the smoothed topic intensity over time
plt.figure(figsize=(16, 8))
for topic in topics_intensity_smoothed.columns:
    plt.plot(topics_intensity_smoothed.index, topics_intensity_smoothed[topic], 
             label=f'Topic {topic}')
# Set labels and titles
plt.xlabel('Year')
plt.ylabel('Topic Probability (Smoothed)')
plt.title('Smoothed Trend of Topic Intensity Over Time (10 years moving average)')
# Set legend
plt.legend(loc='upper right', ncol=2, title='Topics')
plt.grid(True)
# Export visual to be used as Figure 10 of PDF report
plt.savefig('results/Fig10_topic_intensity.pdf')
plt.show()
```
